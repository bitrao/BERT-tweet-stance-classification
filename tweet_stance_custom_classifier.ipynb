# TWEET STANCE CLASSIFIER
### (Custom Model)

```python
import pandas as pd
from models.custom_stance_classifier import CustomStanceClassifier
from utils.data_processor import prepare_data
```

## Load and Prepare Data

```python
train_df = pd.read_csv('data/processed_train.csv')
```

```python
train_df.head()
```

## Modelling

## Bert

```python
bert = 'bert-base-uncased'
```

#### Model

```python
bert_stance_clf = CustomStanceClassifier(bert, 3)
```

#### Prepare train and val dataset

```python
bert_tokenizer = bert_stance_clf.tokenizer
```

```python
bert_train, bert_val, class_weights = prepare_data(train_df, bert_tokenizer, weighted=True)
```

```python
bert_stance_clf.train_model(train_loader=bert_train, val_loader=bert_val, device='cuda', num_epochs=20, class_weights=class_weights)
```
