# TWEET STANCE CLASSIFIER
### (Data Analysis)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load data

```python
train_df = pd.read_csv('data/processed_train.csv')
```

### Examine Target and Stance

```python
train_df.Target.unique()
     
```

```python
train_df.Stance.unique()
```

```python
#Plot number of tweets in each target

sns.set(rc={'figure.figsize':(15,5)})
sns.set_theme(style="darkgrid")
ax = sns.countplot(x="Target",hue='Target' ,data=train_df)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python

#Plot number of tweets in each stance

sns.set(rc={'figure.figsize':(9,5)})
ax = sns.countplot(x='Stance',hue='Sentiment', data=train_df)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
#Plot number of tweets with opinions
sns.set(rc={'figure.figsize':(15,5)})
ax = sns.countplot(x="Target", hue="Stance", data=train_df)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
